Importing pandas and data files

```python
import os, numpy as np, pandas as pd
file = os.sep.join(['seeds_dataset.txt'])
data = pd.read_csv(file, sep="\t", header=None, names=["Area", "Perimeter", "Compactness", "Length of Kernel","Width of Kernel","Asymmetry coefficient","Length of kernel groove","Label"])

#data=data.drop(columns=['h'])
data.head()
```

# Box Plots

```python
for column in data:
    sns.set_style("whitegrid")
    plt.figure(figsize=(10,10))
    sns.boxplot(y=one_data[column],palette='Blues')
    sns.boxplot(y=two_data[column],palette='Greens')
    sns.boxplot(y=three_data[column],palette='Reds')
```

```python
g = sns.PairGrid(data, hue='Label',vars=["Area","Perimeter","Compactness", "Length of Kernel","Width of Kernel","Asymmetry coefficient","Length of kernel groove"])
#g = g.map(plt.scatter)
g = g.map_diag(plt.hist)
g = g.map_offdiag(plt.scatter)
g = g.add_legend()
```

```python
g = sns.PairGrid(data, hue='Label',size=3,vars=["Area","Perimeter","Compactness", "Length of Kernel","Width of Kernel","Asymmetry coefficient","Length of kernel groove"])
g = g.map_upper(plt.scatter, linewidths=1)#, edgecolor="w", s=40)
g = g.map_lower(sns.kdeplot)#, cmap="Blues_d")
g = g.map_diag(plt.hist, lw=3)#, legend=False)
g.savefig('pairgrid.png')
```
